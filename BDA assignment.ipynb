# Transformation

## Data Normalization

```python
df= pd.concat([data_1,df_1,df_2,df_3])
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.head()
```

```python
df.drop(columns="Unnamed: 0",axis=1,inplace=True)
```

```python
df.info()
```

```python
df.isna().sum()
```

## Changing Data Type

```python
df['Date Published']= df['Date Published'].replace(r'(\d+)(st|nd|rd|th)', r'\1', regex=True)
df['Date Published']= pd.to_datetime(df ['Date Published'], format='%d %B %Y')
```

```python
df['Date Flown']= df['Date Flown'].replace(r'(\d+)(st|nd|rd|th)', r'\1', regex=True)
df['Date Flown']= pd.to_datetime(df['Date Flown'], format= '%B %Y')
```

```python
df.info()
```

## Data Cleaning

```python
df.Status.unique()
```

```python
df["Status"].replace({'NotVerified': 'Not Verified', '❎ Unverified': 'Not Verified','❎ Not Verified' : 'Not Verified' ,'Trip Verified':'Verified','Verified Review':'Verified'},inplace = True)
```

```python
df["Status"].fillna(df['Status'].mode()[0],inplace =True)
```

```python
df["Aircraft"].fillna(df["Aircraft"].mode()[0],inplace = True)
```

```python
df.rename(columns={"Date Published":"Date_Published","Travel Type":"Travel_Type","Travel Class":"Travel_Class","Date Flown":"Date_Flown"},inplace = True)
```

```python
df=df.rename(columns={'Seating Comfort':"Seating_Comfort",'Staff Service':'Staff_Service',"Food Quality":"Food_Quality","Ground Service":"Ground_Service","Value for Money":"Value_for_Money","Overall Rating":"Overall_Rating" })
```

```python
df.info()
```

```python
df.dropna(subset="Seating_Comfort",axis= 0,how="any",inplace= True)
```

```python
df.info()
```

```python
df["WiFi"].fillna(df["WiFi"].mean(),inplace = True)
df["Entertainment"].fillna(df["Entertainment"].mean(),inplace = True)
df['Staff_Service'].fillna(df['Staff_Service'].mean(),inplace=True)
df['Food_Quality'].fillna(df['Food_Quality'].mean(),inplace=True)
df['Ground_Service'].fillna(df['Ground_Service'].mean(),inplace=True)
```

```python
df.dropna(subset="Route",axis= 0,how="any",inplace= True)
```

```python
df.info()
```

```python
df.head()
```

## Data Aggregation

```python
agg_df=df.groupby("Travel_Class")['Overall_Rating'].mean().round(2)
```

```python
agg_df
```

## Data Consolidation

```python
df_2= pd.merge(df,agg_df, on='Travel_Class', how='left')
```

```python
df_2.rename(columns={"Overall_Rating_y":"Travel_Class_Rating "},inplace=True)
```

```python
df_2.info()
```

```python
df_2.to_csv("Emirates_Transformed_Data.csv")
```
